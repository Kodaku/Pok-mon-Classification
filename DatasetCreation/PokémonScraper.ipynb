```python
tic = time.time()
# Array di giochi
giochi_names = ['red-blue', 'yellow', 'silver', 'gold', 'crystal', 'ruby-sapphire', 'firered-leafgreen', 'emerald', 'diamond-pearl', 'platinum', 'heartgold-soulsilver',
                'black-white', 'x-y', 'omega-ruby-alpha-sapphire/dex', 'go', 'bank', 'lets-go-pikachu-eevee', 'home' ]

# Get Pokemon names
NO_POKEMON = 151 # Maximum of 802
pokedb_urls = ['https://pokemondb.net/pokedex/'+str(i+1) for i in range(NO_POKEMON)]
pokemon_names = []
for url in pokedb_urls:
    page = bs.BeautifulSoup(requests.get(url).text, 'lxml')
    pokemon_name = page.find('link', {'rel':'canonical'})['href'].split('/')[-1]
    pokemon_names.append(pokemon_name)
  



    
# Write their names to a file for future reference
f = open('pokemon_names', 'w')
f.write(str(pokemon_names))
f.close()

#create a folder for each Pokémon

# Download image of each Pokemon (full colour)
for name in pokemon_names:
  index = 0
  new_path = data_path + "/" + str(pokemon_names.index(name)+1) + "." + name
  try:
    os.mkdir(new_path)
  except:
    pass
  img_url = 'https://img.pokemondb.net/artwork/%s.jpg' % name
  img_bytestr = requests.get(img_url).content
  with open('%s/%s_%d.jpg' % (new_path, name, index), 'wb') as img_f: # Closes nicely
      try:
        img_f.write(img_bytestr)
      except:
        pass
  for gioco in giochi_names:
      index = index + 1
      colored = ""
      if (gioco == 'red-blue' or gioco == 'yellow') :
        colored = "-color"
      #img_url = 'https://img.pokemondb.net/artwork/%s.jpg' % name
      img_url = 'https://img.pokemondb.net/sprites/%s/normal/%s.png' % (gioco, name + colored)
      img_bytestr = requests.get(img_url).content
      with open('%s/%s_%d.jpg' % (new_path, name, index), 'wb') as img_f: # Closes nicely
          try:
            img_f.write(img_bytestr)
          except:
            pass
toc = time.time()
print("Elapsed: ", toc - tic)
```
