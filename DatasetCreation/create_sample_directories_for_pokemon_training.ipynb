```python
import os
from google.colab import drive
import numpy as np
from google.colab.patches import cv2_imshow
import random
```

```python
drive.mount("/content/drive", force_remount=True)
train_path = "/content/drive/Shareddrives/PokémonSerina/Train_Validation_Pokémon_No_Numbers"
test_path = "/content/drive/Shareddrives/PokémonSerina/Test_Pokémon_No_Numbers"
samples_path = "/content/drive/Shareddrives/PokémonSerina/CNN_Train_Samples"
test_samples_path = "/content/drive/Shareddrives/PokémonSerina/CNN_Test_Samples"
```

```python
pokemon_names = os.listdir(train_path)
n_samples = 10
taken_pokemon = []
len(pokemon_names)
```

```python
counter = 0
while counter < 10:
  ok = False
  while not ok:
    index = random.randint(0, len(pokemon_names))
    if pokemon_names[index] not in taken_pokemon:
      taken_pokemon.append(pokemon_names[index])
      ok = True
  counter += 1
taken_pokemon
```

```python
import shutil
```

```python
counter = 0
dir_name = ""
created = False
while not created:
  dir_name = "samples_" + str(counter)
  if not os.path.isdir(samples_path + "/" + dir_name):
    os.mkdir(samples_path + "/" + dir_name)
    created = True
  else:
    counter += 1
dir_name
```

```python
for pokemon in taken_pokemon:
  src = train_path + "/" + pokemon
  dest = samples_path + "/" + dir_name + "/" + pokemon
  shutil.copytree(src, dest, ignore = shutil.ignore_patterns('*.txt', 'a'))
```

```python
counter = 0
dir_name = ""
created = False
while not created:
  dir_name = "samples_" + str(counter)
  if not os.path.isdir(test_samples_path + "/" + dir_name):
    os.mkdir(test_samples_path + "/" + dir_name)
    created = True
  else:
    counter += 1
dir_name
```

```python
for pokemon in taken_pokemon:
  src = test_path + "/" + pokemon
  dest = test_samples_path + "/" + dir_name + "/" + pokemon
  shutil.copytree(src, dest, ignore = shutil.ignore_patterns('*.txt', 'a'))
```
