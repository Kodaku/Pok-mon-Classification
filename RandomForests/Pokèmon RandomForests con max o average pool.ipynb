```python
drive.mount("/content/drive", force_remount=True)
data_path = "/content/drive/Shareddrives/PokémonSerina/Pokemon_resized"
```

```python
dirs = os.listdir(data_path)
sprites=[]


for folder in dirs:
 dirs_2= os.listdir(data_path+'/'+folder)
 label= folder.split('.')
 label= label[1]
 print(label)
 for path in dirs_2:
   sprite= cv2.imread(data_path+'/'+folder+'/'+path)
   sprite = cv2.resize(sprite, (80,80))

   sprite_ridotto = sprite.reshape(1, 80, 80, 3)
   sprite_ridotto = tf.constant(sprite_ridotto, dtype=tf.float64)
   tf.random.set_seed(42)
   #max_average = layers.AveragePooling2D(pool_size=(8,8), strides=8)
   max_average = layers.MaxPooling2D(pool_size=(8,8), strides=8)
   convolved_sprite = max_average(sprite_ridotto)
   new_sprite = tf.squeeze(convolved_sprite)
   new_sprite= tf.cast(new_sprite, dtype=tf.uint8)
   new_sprite = new_sprite.numpy()
   #sprite = np.concatenate((sprite.flatten(),new_sprite.flatten()))
   sprite = new_sprite.flatten()
   arr = [sprite, label]
   sprites.append(arr)


print('- Fine Elaborazione -')
```

```python

# [image, label]

sprites= np.array(sprites, dtype=object)
```

```python
# Python ≥3.5 is required
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"
from sklearn.model_selection import train_test_split

# to make this notebook's output stable across runs
np.random.seed(42)
import random

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#per mescolare le immagini
sprites_permuted=np.random.permutation(sprites)
sprites_permuted[:,1]
```

```python
#controlla le dimensioni delle immagini in sprites
for x in sprites[:,0]:
  print(len(x))
```

```python
flat_arr=[] #input array
targ_arr=[] #output array

for img in sprites_permuted[:,0]:
    flat_arr.append(img)
for img in sprites_permuted[:,1]:
    targ_arr.append(img)

print(flat_arr)
print(targ_arr)
flat_data=np.array(flat_arr)
target=np.array(targ_arr)
df=pd.DataFrame(flat_data)
df['Target']=target
x_data=df.iloc[:,:-1] #input data 
y_data=df.iloc[:,-1] #output data
print(y_data)
```

```python
#prova training e test set 
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(criterion= 'gini', max_leaf_nodes= 20, n_estimators= 200,random_state=42)
print(clf)
```

```python
# fit modello RandomForests
clf.fit(X_train, y_train)
print('The Model is trained well with the given images')
```

```python
y_pred=clf.predict(X_test) #metti clf se no
print(y_pred)
```
